```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('resources/data_interpolated.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python

df = df[df['CityName'] == 'United States']
df.set_index('Date', inplace=True)
df = df[['SalesTotal', 'MORTGAGE30US']]
df['MORTGAGE30US'] = df['MORTGAGE30US'].bfill()
df.dropna(subset=['SalesTotal'], inplace=True)
df.head()
```

```python
# Plot the data
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(df.index, df['MORTGAGE30US'], color='red')
ax2.bar(df.index, df['SalesTotal'], color='blue', alpha = 0.3)



# Set labels and title
ax1.set_xlabel('Time')
```
